# Варіант №3


  \begin{cases}
       8.2x_1 + 7.9x_2 &= 123, \\
        11.2x_1 + 16.3x_2 &= 201, \\
        -16.3x_1 - 8.3x_2 &= 169, \\
   \end{cases}



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# The first equation
x2 = np.linspace(0, 20, 100)
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15, 5), sharey=True, dpi=1080)

f1 = (123 - 7.9 * x2) / 8.2
f2 = (201 - 16.3 * x2) / 11.2
f3 = (169 + 8.3 * x2) / (-16.3)

ax[0].plot(x2, f1, 'r-.')
ax[0].grid(True, linestyle='--', color='gray')
ax[0].set(xlim=(0, 20), ylim=(-20, 30), title='$8.2x_1 + 7.9x_2 = 123$')

ax[1].plot(x2, f2, 'm:')
ax[1].grid(True, linestyle='--', color='gray')
ax[1].set(xlim=(0, 20), title='$11.2x_1 + 16.3x_2 = 201$')

ax[2].plot(x2, f3, color='LightGreen', linestyle='-')
ax[2].grid(True, linestyle='--', color='gray')
ax[2].set(xlim=(0, 20), title='$-16.3x_1 - 8.3x_2 = 169$')

plt.show()
```

```python
# All lines in one figure
x2 = np.linspace(-10, 100, 100)

f1 = (123 - 7.9 * x2) / 8.2
f2 = (201 - 16.3 * x2) / 11.2
f3 = (169 + 8.3 * x2) / (-16.3)

# Налаштовуємо область малювання графіка, її розмір та роздільну здатність
fig, ax = plt.subplots(figsize=(12, 8), dpi=1080)

# Встановлюємо сітку та заголовок графіка
ax.set_title('All lines in one figure')
ax.grid(True, linestyle='--', color='gray')

# Графік 1
ax.plot(x2, f1, 'r-.', label='f[0]')
ax.text(x2[26], f1[25], '$8.2x_1 + 7.9x_2 = 123$', rotation=-np.arctan(7.9/8.2)*180/np.pi + 7, ha='right', va='bottom', color='black')

# Графік 2
ax.plot(x2, f2, 'm:', label='f[1]')
ax.text(x2[47], f2[46], '$11.2x_1 + 16.3x_2 = 201$', rotation=-np.arctan(16.3/11.2)*180/np.pi + 7, ha='right', va='bottom', color='black')

# Графік 3
ax.plot(x2, f3, color='LightGreen', linestyle='-', label='f[2]')
ax.text(x2[25], f3[24], '$-16.3x_1 - 8.3x_2 = 169$', rotation=-np.arctan(-8.3/-16.3)*180/np.pi + 7, ha='right', va='bottom', color='black')

ax.set_xlim(0, 70)
ax.set_ylim(-45, 15)

# Знаходимо точки перетину прямих з використанням numpy.linalg.solve
A12 = np.array([[8.2, 7.9], [11.2, 16.3]])
B12 = np.array([123, 201])
point_1_2 = np.linalg.solve(A12, B12)

A13 = np.array([[8.2, 7.9], [-16.3, -8.3]])
B13 = np.array([123, 169])
point_1_3 = np.linalg.solve(A13, B13)

A23 = np.array([[11.2, 16.3], [-16.3, -8.3]])
B23 = np.array([201, 169])
point_2_3 = np.linalg.solve(A23, B23)

points = np.array([point_1_2, point_1_3, point_2_3])
colors = np.array(['r', 'g', 'b'])

# Позначаємо точки перетину та додаємо стрілочки і пояснення
for i, point in enumerate(points):
    ax.scatter(point[1], point[0], marker='o', color=colors[i], alpha=1, zorder=10)
    ax.annotate(f'Intersection {i + 1}', xy=(point[1], point[0]), xytext=(point[1] + 10, point[0] + 10),
                arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=8, ha='left', va='bottom')


# Сортування точок за зростанням значень x
sorted_points = points[points[:, 1].argsort()]

# Зафарбовуємо трикутник
plt.fill_between(x2, f1, f2, where=(x2 >= sorted_points[0][1]) & (x2 <= sorted_points[1][1] + 1), color='gray', alpha=0.3)
plt.fill_between(x2, f1, f3, where=(x2 >= sorted_points[1][1]) & (x2 <= sorted_points[2][1]), color='gray', alpha=0.3)

lg = ax.legend()

# Збереження у різних форматах
plt.savefig('figure.jpg', format='jpg')
plt.savefig('figure.svg', format='svg')
plt.savefig('figure.png', format='png')

plt.show()
```
